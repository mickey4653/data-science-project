```python

def AgeGroup(age_group):
  
  if (age_group['Age'] < 18):
    return "< 18"
  if (age_group['Age'] >= 18) & (age_group['Age'] < 25):
    return "18-24"
  if (age_group['Age'] >= 25) & (age_group['Age'] < 35):
    return "25-34"
  elif (age_group['Age'] >= 35) & (age_group['Age'] < 45):
    return "35-44"
  elif (age_group['Age'] >= 45) & (age_group['Age'] < 55):
    return "45-54"
  elif (age_group['Age'] >= 55) & (age_group['Age'] < 65):
    return "55-64"
  elif (age_group['Age'] >= 65):
    return "65+"

bwt['age_group_category']= bwt.apply(lambda age_group: AgeGroup(age_group), axis=1 )
```
